```python
import pandas as pd
import numpy as np
```

```python
data1=pd.read_csv('campus.csv')
data1
```

```python
data1.drop(['Timestamp'],axis=1,inplace=True)
```

```python
(data1)
```

```python
data1.rename(columns = {'What is your current year of study':'year', 
                     'What is your current relationship status?': 'relationship',
                     'How do you prefer studying before exams?': 'study',
                     'Do you do competitive programming?': 'CP',
                     'Which hostel do you live in?': 'hostel',
                     'What is your gender?': 'gender',
                     'What is your favourite sport?': 'sport',
                     'Favourite hangout place in campus': 'hangout_spot',
                     'Which state/union territory do you belong to?': 'state',
                     'What is your dept?': 'dept',
                     'What is your preferred coding language?': 'coding_lang',
                     'What is your favourite place to eat at IIT Guwahati?': 'fav_place_to_eat',
                     'Which is your favourite club in IIT Guwahati?': 'fav_club',
                     'Are you satisfied in IIT Guwahati': 'satisfied',
                     'What is your current cpi?': 'cpi'}, inplace = True)

data1.head()
```

```python
print(data1['relationship'].unique())
```

```python
data1['relationship'].replace('In a relationship', 'relationship', inplace=True)
data1['relationship'].replace("It's complicated", 'complicated', inplace=True)

print('Unique values in relationship column after replacements:')
print(data1['relationship'].unique())
```

```python
data1['relationship'].value_counts().sum()
```

```python
data1['relationship'].shape
```

```python
data1['relationship'].fillna(value='Undisclosed',inplace=True)
```

```python
data1.relationship.value_counts()
```

```python
print('Unique values in hangout spot column:')
print(data1['hangout_spot'].unique())
print()

data1['hangout_spot'].replace('IIT G lake (in front of library)', 'IITG_Lake', inplace=True)
data1['hangout_spot'].replace('Serpentine Lake 😏', 'serpentine;)', inplace=True)
data1['hangout_spot'].replace('Lake in front of the hospital', 'Tihor_lake', inplace=True)
data1['hangout_spot'] = data1['hangout_spot'].str.replace(' ', '_')

print('Unique values in hangout spot column after replacements:')
print(data1['hangout_spot'].unique())
```

```python
data1['state'].replace('The Government of NCT of Delhi', 'Delhi', inplace=True)
data1['fav_place_to_eat'].replace('None', np.nan, inplace=True)
data1['fav_club'].replace('ml.ai', 'iitg.ai', inplace=True)
data1['hostel'].replace('Bhramaputra', 'Brahmaputra', inplace=True)
```

```python
def isfloat(x):
    try:
        float(x)
        return True
    except:
        return False
    
for i in data1.index:
    if isfloat(data1.loc[i, 'cpi']):
        data1.loc[i, 'cpi'] = float(data1.loc[i, 'cpi'])
    else:
        data1.loc[i, 'cpi'] = np.nan
        
        # Values of cpi outside range 4-10 is considered fake
    if (data1.loc[i, 'cpi'] >= 10) | (data1.loc[i, 'cpi'] <= 4):
        data1.loc[i, 'cpi'] = np.nan
```
