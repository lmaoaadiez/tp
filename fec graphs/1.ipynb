```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
x=np.arange(3)
y1=[13.74,46.86,19.50]
y2=[19.05,54.57,31.27]
plt.xticks(x,('Thomas Cook','Ease my Trip','ITDC'))
width=0.3
plt.bar(x-width,y1,width=width,label='ROE')
plt.bar(x,y2,width=width,label='ROCE')
plt.legend()
plt.tight_layout()
plt.xlabel('Companies')
plt.ylabel('ROE and ROCE in %')
```

```python
slices=[10087,8027,6038.68]
labels=['Thomas & Cook:10087 Cr','Ease my Trip:8027 Cr','ITDC:6038.68 Cr']
colors=['blue','green','red']
explode=[0.05,0,0]
plt.title('Market Cap in Cr')
plt.pie(slices,labels=labels,colors=colors,explode=explode,shadow=True,startangle=-70,wedgeprops={'edgecolor':'black'})
plt.tight_layout()
```

```python
from matplotlib import patches
import seaborn as sns
sns.set_theme(font_scale = 1.2)
plt.figure(figsize=(10, 10))
plt.title('Market Cap in Cr')
 
patches,texts = plt.pie(
    x=slices, labels=labels,
    colors=sns.color_palette('Set2'),
    startangle=-70,explode=explode,shadow=True,wedgeprops={'edgecolor':'black'},
    labeldistance=0.7)
 
# iterate over the text labels
# make each label bold and center-aligned
for text in texts:
    text.set_fontweight('bold')
    text.set_horizontalalignment('center')
 
```
