```python
import seaborn as sns
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

# Example data
# x_list = [1, 2, 3, 4, 6, 7, 9, 9, 9, 10]
df = pd.DataFrame({"X": cars.weight, "Y": range(len(cars.weight))})

# Create a histplot with KDE
_, FD_bins = np.histogram(cars.weight, bins="fd")
bin_nr = min(len(FD_bins) - 1, 50)
sns.displot(data=df, x="X", bins=bin_nr, stat="density", alpha=0.4, kde=True, kde_kws={"cut": 3})

# Set the title and labels
plt.title("Histogram with Normal Distribution Fit")
plt.xlabel("X values")
plt.ylabel("Density")

plt.show()
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Example data
x_list = cars.weight
df = pd.DataFrame({"X": x_list})

# Create a histogram with KDE
sns.histplot(data=df, x="X", stat="density", kde=True)

# Calculate mean and standard deviation
mean_x = np.mean(x_list)
std_x = np.std(x_list)

# Plot the fitted normal distribution
x_vals = np.linspace(min(x_list), max(x_list), 100)
y_vals = (1 / (std_x * np.sqrt(2 * np.pi))) * np.exp(-(x_vals - mean_x)**2 / (2 * std_x**2))
plt.plot(x_vals, y_vals, label="Fitted Normal Distribution", color="red")

# Set labels and title
plt.xlabel("X values")
plt.ylabel("Density")
plt.title("Histogram with Fitted Normal Distribution")

plt.legend()
plt.show()
```

```python
from scipy import stats
import numpy as np

# Choose some parameters for the skew-normal distribution
a, loc, scale = 5.3, -0.1, 2.2

# Generate a sample of data
data = cars.weight
```

```python
ae, loce, scalee = stats.skewnorm.fit(data)
```

```python
import matplotlib.pyplot as plt

# Plot the histogram of the sample data
plt.hist(data, bins=100, density=True, alpha=0.6, color='g')

# Create an x-axis range for the PDF
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)

# Calculate the PDF of the estimated skew-normal distribution
p = stats.skewnorm.pdf(x, ae, loce, scalee)

# Plot the PDF
plt.plot(x, p, 'k', linewidth=2)
plt.xlabel("Value")
plt.ylabel("Density")
plt.title("Skew-Normal Distribution Fit")
plt.show()
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
cars=sns.load_dataset('mpg').dropna()
```

```python
sns.set_style('whitegrid')
```

```python
cars=cars[cars.cylinders.isin([4,6,8])]
```

```python
sns.violinplot(x=cars.displacement)
```

```python
sns.violinplot(y=cars.displacement,x=cars.cylinders,hue=cars.origin)
```

```python
sns.violinplot(y='displacement',x='cylinders',hue='origin',data=cars[cars.origin.isin(['japan','europe'])],split=True)
plt.legend(loc=2)
```

```python
sns.violinplot(y='displacement',x='cylinders',hue='origin',data=cars[cars.origin.isin(['japan','europe'])],split=True,inner='quartiles')
plt.legend(loc=2)
```

```python
sns.violinplot(y='displacement',x='cylinders',hue='origin',data=cars[cars.origin.isin(['japan','europe'])],split=True,inner='quartiles',density_norm='count')
plt.legend(loc=2)
```

```python
sns.violinplot(y='displacement',x='cylinders',hue='origin',data=cars[cars.origin.isin(['japan','europe'])],split=True,inner='quartiles',density_norm='count',scale_hue=True)
plt.legend(loc=2)
```

```python
sns.violinplot(x=cars.origin,y=cars.displacement,order=['japan','europe','usa'],linewidth=1.5)
```

```python
sns.violinplot(x=cars.cylinders,y=cars.displacement,bw_method=0.2)
```
